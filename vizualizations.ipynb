```python
#Import Libraries
import pandas as pd
from pathlib import Path
import hvplot.pandas
import panel as pn
from panel.interact import interact
from panel import widgets
```

```python
#First visualizations for analysis process
data_path = Path("historical_data.csv")
historical_data = pd.read_csv(data_path, index_col=[0], parse_dates=True, infer_datetime_format=True)
historical_data.head()


df_10yr = historical_data[["10 YR"]]
df_10yr["mmt5"] = (df_10yr - df_10yr.shift(5))
df_10yr["mmt10"] = (df_10yr["10 YR"] - df_10yr["10 YR"].shift(10))
df_10yr["mmt20"] = (df_10yr["10 YR"] - df_10yr["10 YR"].shift(20))
mmt_plot = df_10yr[ "mmt10"].hvplot(title = "Momentum", ylim =[-1.5,1])
```

```python
#Second Visualization for the analyis process - rolling data to get more consistent peaks
mmt_roll = df_10yr["mmt10"].rolling(20).mean().hvplot(title = "20 Day Rolling momentum")
```

```python
#Showing relationship between yield and momentum
yield_mmt_plot = pd.concat([df_10yr["mmt10"].rolling(20).mean(), df_10yr[ "10 YR"]], axis=1).hvplot(title = "10 YR Treasury Yield with Rolling Momentum")
```

```python
# Creating first tab for dashboard - analysis of yields
row1 = pn.Row(mmt_plot, mmt_roll)
row2 = pn.Row(yield_mmt_plot, align="center")
column1 = pn.Column(row1, row2)
column1
```

```python
#Import pickles
returns_df = pd.read_pickle("returns_df.pickle")
sector_df = pd.read_pickle("sector_df.pickle")
```

```python
#Tab 2 for dashboard- individual sector data
sector_returns = ((1+sector_df).cumprod()-1)*100
sector_list = list(sector_returns.columns.levels[0])
sector_dict = {}
for sector in sector_list:
    sector_dict[sector] = sector_returns[sector].hvplot(title = sector, width = 400, xticks = [0,15,30,45,60,75,90], xlim=[0,90], ylabel = "Cumulative Percent Return", shared_axes = False)

row4 = pn.Row(sector_dict['XLB'], sector_dict['XLE'], sector_dict['XLE'])
row5 = pn.Row(sector_dict['XLI'], sector_dict['XLK'], sector_dict['XLK'])
row6 = pn.Row(sector_dict['XLU'], sector_dict['XLV'], sector_dict['XLY'])
column2 = pn.Column(row4,row5,row6)
column2
```

```python
sector_average_returns_plot = returns_df.hvplot(title="Average Returns for Each Sector", width=1000, height=500, xticks = [0,15,30,45,60,75,90], xlim=[0,90], ylabel = "Cumulative Percent Return", shared_axes = False )
sector_average_returns_plot
```

```python
#Tab 3 for dashboard - average sector returns
row3 = pn.Row(sector_average_returns_plot)
row3
```

```python
#Create Tabs
dashboard = pn.Tabs(('Analysis', column1),
                   ('Individual Sector Performance', column2),
                   ('Average Sector Performance', row3))
dashboard
```
